```python
import pandas as pd 
import numpy as np
import math as m 
```

# Goal
find the distance between hotel cordinate 40.7522, -73.9795 and all airbnb locations

```python
#import data into data frame
airbnb = pd.read_csv("Airbnb.csv")
airbnb = airbnb[['name','host_id','latitude','longitude']]
airbnb
```

```python
# transform list intp data frame
 

def distance():
    data = pd.read_csv("Airbnb.csv", usecols = ['host_id', 'latitude', 'longitude'])
    lat = data['latitude']
    lon = data['longitude']
    dist = []
    radius = 6371
    
    
    for i in range(len(lat)):
        phi1 = m.radians(40.7522)
        phi2 = m.radians(lat[i])
        lambda1 = m.radians(-73.9795)
        lambda2 = m.radians(lon[i])
        dlat = phi2 - phi1
        dlon = lambda2 -lambda1
        a = m.sqrt(m.sin(dlat/2)**2+m.cos(phi1)*m.cos(phi2)*m.sin(dlon/2)**2) #use of Haversine formula
        d = 2*radius*np.arcsin(a)
        dist.append(d)
    return dist

distance()

distance = distance()

airbnb['Distance_to_Library_Hotel'] = distance
airbnb
```

# Filter to show only airbnbs under 500m

```python
airbnb = airbnb[airbnb.Distance_to_Hotel < 0.5 ] 
airbnb
```
