# Filters
- we could limit properties only available for more the (100) number of days
- filter for only entire home/apt

```python
entire = df3[df3.room_type == 'Entire home/apt']
man = entire[entire.neighbourhood_group == 'Manhattan']
man1 = man[man.availability_365 > 100]
man1
```

# Filter for neighbourhoods
- group by function 
- collapse the data to give the average price per neighberhood

```python
# Mean price per neighberhood
nh = pd.DataFrame(man1.groupby('neighbourhood')['price'].aggregate(np.mean))
nh
```

# Date filters

we need to find the first listing for each location 
- USE df 1 and merge on listing id
- RENAME DF1 to ID

```python
df1 = df1.rename(columns={'listing_id': 'id'})
df1['date'] = pd.to_datetime(df1['date'])
df1 = pd.DataFrame(df1.groupby('id')['date'].min()) #min dates for each listing
```

```python
DF = pd.merge(man1, df1, on = 'id')
DF
```
